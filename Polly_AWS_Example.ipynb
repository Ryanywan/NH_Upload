# AWS TTS
## AWS Cloud에 연결하는 library - boto3를 호출합니다.

```python
import boto3
```

## Session을 통해 access key와 secret key를 입력하고 필요한 서비스를 호출합니다.

```python
####################### boto3를 이용하여 AWS cloud에 접속합니다. (이메일 로그인과 비슷한 방식) #######################
# aws_access_key_id : 아이디
# aws_secret_access_key: 패스워드
# region_name: 호출 지역
# client: 호출서비스

polly_client = boto3.Session(
                aws_access_key_id= #개인 필수정보 1                      
    aws_secret_access_key= #개인 필수정보 2
    region_name='ap-northeast-2').client('polly') #사용 지역과 서비스 입력


####################### polly 서비스에 #######################
# response: 호출하는 내용 parameter 값 설정!
# VoiceId : 목소리 (210813 기준: 'Seoyeon'이 한국어 최적화)
# OutputFormat: mp3 음성
# Text: 리턴하고자 하는 값

response = polly_client.synthesize_speech(VoiceId='Seoyeon',
                OutputFormat='mp3', 
                Text = '안녕하세요. AI 융합센터 입니다. 많은 관심 부탁드립니다.',
                Engine = 'neural')

file = open('speech2.mp3', 'wb')
file.write(response['AudioStream'].read())
file.close()
```
